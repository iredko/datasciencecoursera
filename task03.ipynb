```python
import numpy
import matplotlib.pyplot as plt
from IPython.display import Image
```

Importing data

```python
data = numpy.genfromtxt('bkgd.csv', skip_header=1).T
```

1D histogram

```python
plt.subplot(1, 2, 1)
n, bins, patches = plt.hist(data[0], bins=30, color="b", histtype="step")
plt.subplot(1, 2, 2)
plt.hist(data[0], bins=30, color="b",histtype="step")
plt.xlim((0, 0.5))
plt.xlabel("M2AB")
plt.ylabel("N")
plt.errorbar(numpy.array([(bins[i] + bins[i + 1]) / 2 for i in range(len(bins) - 1)]), n, yerr=numpy.sqrt(n), fmt='.', color='r')
plt.show()
```

```python
Image(filename='./figure_1.png')
```

2D histogram

```python
plt.hist2d(data[0],data[1],bins=30,cmap="Blues")
plt.colorbar()
plt.xlabel("M2AB")
plt.ylabel("M2AC")
plt.show()
```

```python
Image(filename='./figure_2.png')
```

3D histogram

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.view_init(75, 30)

X = data[0]	
Y = data[1]
Z, X, Y = numpy.histogram2d(X, Y, bins=30)
X, Y = numpy.meshgrid(X, Y)
X = numpy.array([numpy.array(x[1:]) for x in X[1:]])
Y = numpy.array([numpy.array(y[1:]) for y in Y[1:]])
X.flatten()
Y.flatten()
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='Blues',
        linewidth=0, antialiased=True)


fig.colorbar(surf, shrink=0.5, aspect=5)
plt.minorticks_on()

plt.show()
```

```python
Image(filename='./figure_3.png')
```

Pearson correlation table

```python
from scipy.stats.stats import pearsonr
pearson = numpy.array([numpy.array([pearsonr(arr_x,arr_y)[0] for arr_x in data[:9]]) for arr_y in data[:9]])
fig=plt.matshow(pearson, cmap="Reds")
ticks=["M2AB", "M2AC", "XA", "XB", "XC", "Y1", "Y2", "Y3", "Y4"]
plt.xticks(numpy.arange(9),ticks,rotation=90)
plt.yticks(numpy.arange(9),ticks)
plt.gca().invert_yaxis()
plt.show()
```
